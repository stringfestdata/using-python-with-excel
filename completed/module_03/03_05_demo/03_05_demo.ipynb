```python
import pandas as pd
import seaborn as sns
import openpyxl
import matplotlib.pyplot as plt

mpg = sns.load_dataset('mpg')
mpg.head()

wb = openpyxl.Workbook()
```

```python
ws1 = wb.create_sheet('violin')
```

```python
sns.violinplot(x='origin', y='mpg', data=mpg)
```

```python
plt.savefig('violin.jpg')
```

```python
img = openpyxl.drawing.image.Image('violin.jpg')
img.anchor = 'A1'
ws1.add_image(img)
```

```python
ws1 = wb.create_sheet('strip')

sns.stripplot(x='origin', y='mpg', data=mpg)
plt.savefig('strip.jpg')

img = openpyxl.drawing.image.Image('strip.jpg')
img.anchor = 'A1'
ws1.add_image(img)
```

```python
ws1 = wb.create_sheet('swarm')

sns.swarmplot(x='origin', y='mpg', data=mpg)
plt.savefig('swarm.jpg')

img = openpyxl.drawing.image.Image('swarm.jpg')
img.anchor = 'A1'
ws1.add_image(img)
```

```python
wb.save('python-viz.xlsx')
```
